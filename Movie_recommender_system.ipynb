- Here we are taking the name values and appending them to the empty list
- we are first converting the string to the list by using the literal evaluation.

```python
def convert(obj):
  L=[]
  for i in ast.literal_eval(obj):
    L.append(i['name'])
  return L
```

```python
movies['genres']=movies['genres'].apply(convert)
```

```python
movies.head()
```

```python
def convert3(obj):
  L=[]
  counter=0
  for i in ast.literal_eval(obj):
    if counter != 3:
      L.append(i['name'])
      counter += 1
    else:
      break
  return L
```

```python
movies['cast']= movies['cast'].apply(convert3)
```

```python
movies['keywords']= movies['keywords'].apply(convert3)
```

- We need to find where the job is director
- we need to display only the director of the movies

```python
def fetch_director(obj):
  L=[]
  for i in ast.literal_eval(obj):
    if i['job']=='Director':
      L.append(i['name'])
  return L
```

```python
movies['crew']=movies['crew'].apply(fetch_director)
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
['love','loves','loving','loved']
```

```python
def stem(text):
  y=[]
  for i in text.split():
    y.append(ps.stem(i))
  return " ".join(y)
```

```python
%%writefile app.py
import streamlit as st
import pickle
import requests

movies=pickle.load(open('movie_list.pkl','rb'))
similarity=pickle.load(open('similarity.pkl','rb'))

def fetch_poster(movie_id):
  url = "https://api.themoviedb.org/3/movie/{}?api_key=97f21ae5d6cd22f8a393f78c87b3ba35&language=en-US".format(movie_id)
  data = requests.get(url)
  data = data.json()
  poster_path = data['poster_path']
  full_path='https://image.tmdb.org/t/p/w500/'+poster_path
  return full_path

def recommend(movie):
  index=movies[movies['title']==movie].index[0]
  distances=sorted(list(enumerate(similarity[index])), key=lambda x:x[1], reverse=True)
  recommended_movie_names=[]
  recommended_movie_posters=[]
  for i in distances[1:11]:
    movie_id = movies.iloc[i[0]].movie_id
    recommended_movie_names.append((movies.iloc[i[0]].title))
    recommended_movie_posters.append(fetch_poster(movie_id))
  return recommended_movie_names,recommended_movie_posters

st.header("Movie Recommender System")
movie_list = movies['title'].values
selected_movie=st.selectbox("Type or select the movie from dropdown", movie_list, index=None)

if st.button("Show Recommendations"):
  recommended_movie_names,recommended_movie_posters=recommend(selected_movie)

  st.subheader("Below are the recommended movies")

  c0,c1,c2,c3,c4,c5,c6,c7,c8,c9=st.tabs(["First","Second","Third","Fourth","Fifth","Sixth","Seventh","Eighth","Nineth","Tenth"])
  with c0:
      st.text(recommended_movie_names[0])
      st.image(recommended_movie_posters[0])
  with c1:
      st.text(recommended_movie_names[1])
      st.image(recommended_movie_posters[1])
  with c2:
      st.text(recommended_movie_names[2])
      st.image(recommended_movie_posters[2])
  with c4:
      st.text(recommended_movie_names[3])
      st.image(recommended_movie_posters[3])
  with c5:
      st.text(recommended_movie_names[4])
      st.image(recommended_movie_posters[4])
  with c6:
      st.text(recommended_movie_names[5])
      st.image(recommended_movie_posters[5])
  with c7:
      st.text(recommended_movie_names[6])
      st.image(recommended_movie_posters[6])
  with c8:
      st.text(recommended_movie_names[7])
      st.image(recommended_movie_posters[7])
  with c9:
      st.text(recommended_movie_names[8])
      st.image(recommended_movie_posters[8])




```
